# File Handling Functions:


```python
import csv

# Example 1: Writing to a CSV file
fields = ['Name', 'Email']
rows = [['Alice', 'alice@example.com'],
        ['Bob', 'bob@example.com'],
        ['Charlie', 'charlie@example.com']]

filename = 'contacts.csv'

with open(filename, 'w', newline='') as csvfile:
    csvwriter = csv.writer(csvfile)
    csvwriter.writerow(fields)
    csvwriter.writerows(rows)
```

```python
# Example 2: Reading from a CSV file
with open(filename, 'r') as csvfile:
    csvreader = csv.reader(csvfile)
    fields = next(csvreader)  # Reading header (first row)
    for row in csvreader:
        print(row)
```

```python
# Example 3: Appending to a CSV file
new_rows = [['David', 'david@example.com'], ['Eva', 'eva@example.com']]

with open(filename, 'a', newline='') as csvfile:
    csvwriter = csv.writer(csvfile)
    csvwriter.writerows(new_rows)
```

```python
# Example 4: Reading and writing text files
# Writing to a text file
with open('example.txt', 'w') as f:
    f.write("This is an example text file.\n")
    f.write("It contains several lines.\n")

# Reading from a text file
with open('example.txt', 'r') as f:
    content = f.read()
    print(content)
```

```python
# Example 5: Appending to a text file
with open('example.txt', 'a') as f:
    f.write("Adding another line to the text file.\n")
```

# Pandas

```python
import pandas as pd

# Creating a DataFrame from a dictionary
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 35],
        'Score': [85.5, 92.7, 78.1]}
df = pd.DataFrame(data)
print(df)

df.to_csv('example.csv', index=False)
```

```python
# Reading a CSV file into a DataFrame
df = pd.read_csv('example.csv')
print(df)
```

```python
# Concatenating two DataFrames
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})
result = pd.concat([df1, df2], axis=0)  # Concatenating along rows
print(result)
```

```python
import pandas as pd

# Creating a DataFrame
df = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})

# New row to append
new_row = pd.DataFrame({'A': [5], 'B': [6]})

# Using pd.concat to append the new row
df = pd.concat([df, new_row], ignore_index=True)

print(df)
```

```python
# Transposing a DataFrame
df = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
print(df.T)
```

```python
# Dropping columns and rows from a DataFrame
df = pd.DataFrame({'A': [1, 2], 'B': [3, 4], 'C': [5, 6]})
df = df.drop(['B'], axis=1)  # Dropping a column
df = df.drop([0], axis=0)    # Dropping a row
print(df)
```

```python
# Resetting the index of a DataFrame
df = pd.DataFrame({'A': [1, 2], 'B': [3, 4]}, index=['x', 'y'])
df = df.reset_index()  # Resetting the index
print(df)
```

```python
# Splitting a DataFrame using sample()
df = pd.DataFrame({'A': range(10), 'B': range(10, 20)})
train = df.sample(frac=0.7, random_state=42)  # 70% for training
test = df.drop(train.index)  # Remaining 30% for testing
print(train)
print(test)
```

```python
# Getting a summary of statistics for the DataFrame
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
print(df.describe())
```

```python
# Selecting data based on index positions
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
print(df.iloc[0])  # First row
print(df.iloc[:, 1])  # Second column
```

```python
# Selecting data based on labels
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]}, index=['a', 'b', 'c'])
print(df.loc['a'])  # Row with label 'a'
print(df.loc[:, 'B'])  # Column 'B'
```

```python
# Dropping rows with missing values
df = pd.DataFrame({'A': [1, None, 3], 'B': [4, 5, None]})
clean_df = df.dropna()  # Drops rows where any NaN is present
print(clean_df)
```

```python
# Dropping duplicate rows
df = pd.DataFrame({'A': [1, 2, 2], 'B': [3, 4, 4]})
df = df.drop_duplicates()
print(df)
```

```python
# Calculating correlation between columns
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]})
correlation_matrix = df.corr()
print(correlation_matrix)
```

```python
# Grouping data into bins
df = pd.DataFrame({'Age': [20, 35, 50, 65, 80]})
df['AgeGroup'] = pd.cut(df['Age'], bins=[0, 30, 50, 100], labels=['Young', 'Middle-Aged', 'Senior'])
print(df)
```

```python
# Creating a new column based on existing data
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
df['C'] = df['A'] + df['B']  # New column 'C' based on the sum of 'A' and 'B'
print(df)
```

```python
# Getting the shape (number of rows and columns) of the DataFrame
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
print(df.shape)  # Outputs the shape as (number of rows, number of columns)
```

# NumPy Operations:

```python
import numpy as np

# Generating a 3x3 matrix of random numbers drawn from a normal distribution
random_matrix = np.random.randn(3, 3)
print(random_matrix)
```

```python
import numpy as np

# Creating an array that includes infinity
data = np.array([1, 2, np.inf, -np.inf, 3, 4])
print(data)

# Handling infinity in computations
# For example, replacing infinity values with NaN (Not a Number)
data[np.isinf(data)] = np.nan
print(data)
```

# Data Visualization with Matplotlib:

```python
import pandas as pd
import matplotlib.pyplot as plt

# Creating a simple DataFrame
data = {'Year': [2018, 2019, 2020, 2021],
        'Sales': [100, 120, 140, 160]}
df = pd.DataFrame(data)

# Plotting the DataFrame
df.plot(x='Year', y='Sales', kind='line')  # Line plot of 'Sales' over 'Year'

# Showing the plot
plt.show()
```
