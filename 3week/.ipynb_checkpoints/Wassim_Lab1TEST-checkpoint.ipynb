# 0 Import

```python
import pandas as pd
```

# 1 Load the Dataset
## Load the nba.csv file into a DataFrame.

```python
df = pd.read_csv('nba.csv')
```

# 2 Process the data

## 2.1 Remove any rows with missing information

```python
clean_df = df.dropna()
clean_df
```

## 2.2 Remove all duplicate rows

```python
nodup_clean_df = clean_df.drop_duplicates()
nodup_clean_df
```

## 2.3 Shuffle the data and split it into training (80%) and test (20%) sets.

```python
train_df = nodup_clean_df.sample(frac=0.8, random_state=0)
print(train_df)
test_df = nodup_clean_df.drop(train_df.index)
print(test_df)
```

# 3 Statistical Description
## Calculate and display the statistical description for both the training and test datasets.

```python
print(train_df.describe())
print(test_df.describe())
```

# 4 Correlation Analysis

## 4.1 Create the dataframes

```python
corr_df_Boston = nodup_clean_df[nodup_clean_df['Team']=='Boston Celtics']['Salary'].reset_index(drop=True)
print(corr_df_Boston)
corr_df_Brooklyn = nodup_clean_df[nodup_clean_df['Team']=='Brooklyn Nets']['Salary'].reset_index(drop=True)
print(corr_df_Brooklyn)
corr_df_NY = nodup_clean_df[nodup_clean_df['Team']=='New York Knicks']['Salary'].reset_index(drop=True)
print(corr_df_NY)
```

## 4.2 Concatane the dataframes

```python
corr_df_concat = pd.concat([corr_df_Boston, corr_df_Brooklyn, corr_df_NY], axis=1)
corr_df_concat.columns = ['Boston', 'Brooklyn', 'NY']
corr_df_concat
```

## 4.3 Find the correlation between the salaries of the teams

```python
corr_df_concat = corr_df_concat.dropna() # clean the NaN rows
corr_df = corr_df_concat.corr()
corr_df
```
