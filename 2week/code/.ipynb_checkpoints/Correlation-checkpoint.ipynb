```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

# Explanation:
# Pearson Correlation measures the linear relationship between two continuous variables.
# The coefficient (r) ranges from -1 to 1:
# - r = 1: Perfect positive linear correlation.
# - r = -1: Perfect negative linear correlation.
# - r = 0: No linear correlation.

# Set random seed for reproducibility
np.random.seed(0)

# 1. Generate sample data for each type of correlation:
# Positive correlation:
x_pos = np.linspace(0, 10, 100)
y_pos = 2 * x_pos + np.random.normal(0, 1, 100)  # y = 2x + some noise

# Negative correlation:
x_neg = np.linspace(0, 10, 100)
y_neg = -2 * x_neg + np.random.normal(0, 1, 100)  # y = -2x + some noise

# No correlation:
x_none = np.linspace(0, 10, 100)
y_none = np.random.normal(0, 10, 100)  # Random noise, no relation to x

# 2. Calculate Pearson Correlation for each dataset
r_pos, _ = pearsonr(x_pos, y_pos)
r_neg, _ = pearsonr(x_neg, y_neg)
r_none, _ = pearsonr(x_none, y_none)

# 3. Create the figure with three subplots (one row, three columns)
plt.figure(figsize=(18, 5))

# Plot 1: Positive Correlation
plt.subplot(1, 3, 1)
plt.scatter(x_pos, y_pos, color='dodgerblue', s=100, alpha=0.7)
plt.plot(x_pos, 2 * x_pos, color='blue', linestyle='-', linewidth=2)
plt.title(f'Positive Correlation (r = {r_pos:.2f})')

# Plot 2: Negative Correlation
plt.subplot(1, 3, 2)
plt.scatter(x_neg, y_neg, color='tomato', s=100, alpha=0.7)
plt.plot(x_neg, -2 * x_neg, color='red', linestyle='-', linewidth=2)
plt.title(f'Negative Correlation (r = {r_neg:.2f})')

# Plot 3: No Correlation
plt.subplot(1, 3, 3)
plt.scatter(x_none, y_none, color='limegreen', s=100, alpha=0.7)
plt.title(f'No Correlation (r = {r_none:.2f})')

# Show the plots
plt.tight_layout()
plt.show()

# Explanation of Pearson Correlation:
# - Positive correlation (r ~ 1): Indicates that as x increases, y increases in a linear relationship.
# - Negative correlation (r ~ -1): Indicates that as x increases, y decreases in a linear relationship.
# - No correlation (r ~ 0): Indicates no linear relationship between x and y.
```
