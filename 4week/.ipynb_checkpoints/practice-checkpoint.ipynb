# Import

```python
import pandas as pd
```

# 1. Read csv file

```python
nba_df = pd.read_csv('nba.csv')
nba_df.head()
```

# 2. Count missing values

```python
nba_df.isnull().sum()
```

# 3. Remove all row with missing values

```python
nba_df_clean = nba_df.dropna()
nba_df_clean
```

# 4. Convert csv to txt

```python
nba_df_clean.to_csv('nba_df_clean.txt', sep=',', index=False)
```

# 5. Read first 1024 bytes

```python
with open('nba_df_clean.txt') as nba:
    print(nba.read(1024))
```

# 6. Group by Team and calculate average Salary

```python
nba_team_grp_salaries = nba_df_clean.groupby('Team')['Salary'].mean().round(2).reset_index()
nba_team_grp_salaries
```

# 7. Group by Position and calcul mean Age

```python
nba_age_grp = nba_df_clean.groupby('Position')['Age'].mean()
nba_age_grp
```

# 8. Split dataframe in half and concatane it back

```python
# IT'S DROPS SO IT DROPS FROM THE SELECTED INDEX, NOT KEEP!
nba_1st_half = nba_df_clean.drop(nba_df_clean.iloc[len(nba_df_clean)//2:].index)
nba_2nd_half = nba_df_clean.drop(nba_1st_half.index)
nba_concat = pd.concat([nba_1st_half, nba_2nd_half], axis=0)
nba_concat
```

# 9. Sort dataframe by Salary in descending order

```python
nba_sorted = nba_df_clean.sort_values('Salary', ascending=False)
nba_sorted.head(10)
```

# 10. Count number of non-null values

```python
nba_df.count()
```

# 11. Create categories for Age

```python
nba_age_categories = pd.cut(nba_df['Age'], bins=[20, 25, 30, 35], labels=['Young', 'Middle-aged', 'Senior'])
nba_df['Age Category'] = nba_age_categories
nba_df.head()
```
