**Linear Regression Example**

```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

# Sample data (X: features, y: target)
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([1, 2, 3, 3.5, 4.5])

# Create and train the model
model = LinearRegression()
model.fit(X, y)

# Make predictions
y_pred = model.predict(X)

# Visualize the results
plt.scatter(X, y, color='blue', label='Actual data')
plt.plot(X, y_pred, color='red', label='Predicted line')
plt.xlabel('Feature X')
plt.ylabel('Target y')
plt.legend()
plt.show()
```

**Ridge Regression Example**

```python
from sklearn.linear_model import Ridge

# Sample data
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([1, 2, 3, 3.5, 4.5])

# Create Ridge model with regularization parameter alpha
ridge_model_1 = Ridge(alpha=10)
ridge_model_1.fit(X, y)

ridge_model_2 = Ridge(alpha=0)
ridge_model_2.fit(X, y)

# Make predictions
y_pred_ridge_1 = ridge_model_1.predict(X)
y_pred_ridge_2 = ridge_model_2.predict(X)

# Plot the predictions
plt.scatter(X, y, color='blue', label='Actual data')
plt.plot(X, y_pred_ridge_1, color='green', label='Ridge Regression line alpha = 10')
plt.plot(X, y_pred_ridge_2, color='red', label='Linear Regression') # alpha = 0 SAME AS LINEAR REGRESSION
plt.xlabel('Feature X')
plt.ylabel('Target y')
plt.legend()
plt.show()
```

**Lasso Regression Example**

```python
from sklearn.linear_model import Lasso

# Create Lasso model with regularization parameter alpha
lasso_model_1 = Lasso(alpha=10)
lasso_model_1.fit(X, y)

lasso_model_2 = Lasso(alpha=0)
lasso_model_2.fit(X, y)

# Make predictions
y_pred_lasso_1 = lasso_model_1.predict(X)
y_pred_lasso_2 = lasso_model_2.predict(X)

# Plot the predictions
plt.scatter(X, y, color='blue', label='Actual data')
plt.plot(X, y_pred_lasso_1, color='orange', label='Lasso Regression line alpha = 10')
plt.plot(X, y_pred_lasso_2, color='purple', label='Linear Regression')
plt.xlabel('Feature X')
plt.ylabel('Target y')
plt.legend()
plt.show()

# The error message is just telling us that with alpha = 0, it's better to use linear regression
```

**Elastic Net Regression Example**

```python
from sklearn.linear_model import ElasticNet

# Create Elastic Net model with both L1 and L2 regularization
elastic_model = ElasticNet(alpha=0.1, l1_ratio=0.5) # hyperparameters: must be chose before training the model
elastic_model.fit(X, y)

# Make predictions
y_pred_elastic = elastic_model.predict(X)

# Plot the predictions
plt.scatter(X, y, color='blue', label='Actual data')
plt.plot(X, y_pred_elastic, color='purple', label='Elastic Net Regression line')
plt.xlabel('Feature X')
plt.ylabel('Target y')
plt.legend()
plt.show()
```
