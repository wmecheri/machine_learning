# 1. K-Nearest Neighbors (KNN)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

# Data: two classes with slight overlap
X = np.array([[1, 2], [2, 3], [3, 3], [6, 8], [7, 7], [8, 6]])
y = np.array([0, 0, 0, 1, 1, 1])

# KNN classifier with 3 neighbors
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X, y)

# Visualization
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', label='Data Points')
new_point = np.array([[5, 5]])
predicted_class = knn.predict(new_point)
plt.scatter(new_point[:, 0], new_point[:, 1], c=predicted_class, s=100, marker='x', label='New Point')
plt.legend()
plt.title("K-Nearest Neighbors Classification")
plt.show()
```

# 2. K-Means Clustering

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Generate dataset with 3 centers
X, _ = make_blobs(n_samples=300, centers=3, cluster_std=0.6, random_state=0)

# K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=0)
y_kmeans = kmeans.fit_predict(X)

# Visualization
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis', label='Clustered Points')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', marker='*', label='Centroids')
plt.legend()
plt.title("K-Means Clustering")
plt.show()
```

# 3. DBSCAN (Density-Based Spatial Clustering of Applications with Noise)

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons

# Generate moon-shaped data
X, _ = make_moons(n_samples=300, noise=0.1)

# DBSCAN clustering
dbscan = DBSCAN(eps=0.2, min_samples=5)
y_dbscan = dbscan.fit_predict(X)

# Visualization
plt.scatter(X[:, 0], X[:, 1], c=y_dbscan, cmap='viridis')
plt.title("DBSCAN Clustering")
plt.show()
```

# 4. Gaussian Mixture Modeling (GMM)

```python
from sklearn.mixture import GaussianMixture

# Generate synthetic data with 3 clusters
X, _ = make_blobs(n_samples=300, centers=3, cluster_std=1.0, random_state=42)

# GMM clustering
gmm = GaussianMixture(n_components=3, random_state=0)
y_gmm = gmm.fit_predict(X)

# Visualization
plt.scatter(X[:, 0], X[:, 1], c=y_gmm, cmap='viridis', label='Clustered Points')
plt.scatter(gmm.means_[:, 0], gmm.means_[:, 1], s=300, c='red', marker='*', label='Gaussian Centers')
plt.legend()
plt.title("Gaussian Mixture Modeling (GMM)")
plt.show()
```

# 5. Hierarchical Clustering

```python
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

# Generate dataset
X, _ = make_blobs(n_samples=20, centers=3, cluster_std=0.8, random_state=42)

# Hierarchical clustering (linkage)
linked = linkage(X, method='ward')

# Plot dendrogram
plt.figure(figsize=(8, 6))
dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.title("Hierarchical Clustering Dendrogram")
plt.show()

# Agglomerative clustering with 3 clusters
hierarchical = AgglomerativeClustering(n_clusters=3, metric='euclidean', linkage='ward')
y_hierarchical = hierarchical.fit_predict(X)

# Visualization
plt.scatter(X[:, 0], X[:, 1], c=y_hierarchical, cmap='viridis')
plt.title("Agglomerative Clustering with 3 Clusters")
plt.show()
```
